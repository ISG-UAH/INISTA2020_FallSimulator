```python
import os
import sys
sys.path
sys.path.append("/usr/lib/jvm/java-11-openjdk-amd64/bin/")
os.environ["JAVA_HOME"] = "/usr/lib/jvm/java-11-openjdk-amd64/"

import weka.core.jvm as jvm
jvm.start(packages=True)
```

```python
data_dir = '/media/animal/DATA/Linux/armando_ml/2020_electronics/2020-electronics/data/arff/'
os.listdir(data_dir)
```

```python
import weka.core.converters as converters


up_sdwd = converters.load_any_file(os.path.join(data_dir, 'upfall_standardized_balanced_sdwd.arff'), class_index = 'last')
up_fwd_knees = converters.load_any_file(os.path.join(data_dir, 'upfall_standardized_balanced_fwd_knees.arff'), class_index = 'last') 
up_fwd_hands = converters.load_any_file(os.path.join(data_dir, 'upfall_standardized_balanced_fwd_hands.arff'), class_index = 'last') 
up_bwd = converters.load_any_file(os.path.join(data_dir, 'upfall_standardized_balanced_bwd.arff'), class_index = 'last') 
uma_lateral = converters.load_any_file(os.path.join(data_dir, 'umafall_standardized_balanced_lateral.arff'), class_index = 'last') 
uma_fwd = converters.load_any_file(os.path.join(data_dir, 'umafall_standardized_balanced_fwd.arff'), class_index = 'last') 
uma_bwd = converters.load_any_file(os.path.join(data_dir, 'umafall_standardized_balanced_bwd.arff'), class_index = 'last') 
sim = converters.load_any_file(os.path.join(data_dir, 'simulated_standardized.arff'), class_index = 'last') 
```

```python
uma_trainers = [
 'umafall_standardized_balanced_lateral.arff',
 'umafall_standardized_balanced_fwd.arff',
 'umafall_standardized_balanced_bwd.arff',
]

up_trainers = [
    'upfall_standardized_balanced_sdwd.arff',
 'upfall_standardized_balanced_fwd_knees.arff',
 'upfall_standardized_balanced_fwd_hands.arff',
 'upfall_standardized_balanced_bwd.arff',
]
```

```python
REDUCED_PRINT = True

import pandas as pd

resumen_df = pd.DataFrame(columns=['TrainingSet','C4.5','PART','MLP','AdaBoost + DecisionStump','SVM','Random Forest','Class'])

uma_trainers = [
 'umafall_standardized_balanced_lateral.arff',
 'umafall_standardized_balanced_fwd.arff',
 'umafall_standardized_balanced_bwd.arff',
]

up_trainers = [
    'upfall_standardized_balanced_sdwd.arff',
    'upfall_standardized_balanced_fwd_knees.arff',
    'upfall_standardized_balanced_fwd_hands.arff',
    'upfall_standardized_balanced_bwd.arff',
]

j48_uma = ['-C', '0.01', '-M', '3', '-O']
j48_up = ['-C', '0.01', '-M', '2', '-O', '-J']
j48_base = 'weka.classifiers.trees.J48'


part_uma = ['-C', '0.11', '-M', '3']
part_up = ['-C', '0.26', '-M', '4']
part_base = 'weka.classifiers.rules.PART'

mlp_uma = ['-L', '0.05', '-M', '0.05', '-N', '50', '-V', '0', '-S', '0', '-E', '20', '-H', 'i', '-D']
mlp_up = ['-L', '0.05', '-M', '0.4', '-N', '50', '-V', '0', '-S', '0', '-E', '20', '-H', 't', '-D']
mlp_base = 'weka.classifiers.functions.MultilayerPerceptron'

ada_base = 'weka.classifiers.meta.AdaBoostM1'
ada_opts = ['-P' ,'100' ,'-S' ,'1' ,'-I' ,'10' ,'-W' ,'weka.classifiers.trees.DecisionStump']

svm_base = 'weka.classifiers.functions.LibSVM'
svm_opts = ['-S', '0', '-K', '2', '-D', '3', '-G', '0.0', '-R', '0.0', '-N', '0.5', '-M', '40.0', '-C', '1.0', '-E',' 0.001', '-P', '0.1', '-model', '/home/animal', '-seed', '1']

randomforest_base = 'weka.classifiers.trees.RandomForest'
randomforest_opts = ['-P', '100', '-I', '100', '-num-slots', '1', '-K', '0', '-M', '1.0', '-V', '0.001', '-S', '1']

from weka.classifiers import Classifier
from weka.classifiers import Evaluation


classifiers_uma = [
    ('C4.5',j48_base, j48_uma),
    ('PART',part_base, part_uma),
    ('MLP',mlp_base, mlp_uma),
    ('AdaBoost + DecisionStump',ada_base, ada_opts),
    ('SVM',svm_base, svm_opts),
    ('Random Forest',randomforest_base, randomforest_opts),
]

classifiers_up = [
    ('C4.5',j48_base, j48_up),
    ('PART',part_base, part_up),
    ('MLP',mlp_base, mlp_up),
    ('AdaBoost + DecisionStump',ada_base, ada_opts),
    ('SVM',svm_base, svm_opts),
    ('Random Forest',randomforest_base, randomforest_opts),
]

for i, df in enumerate([uma_lateral, uma_fwd, uma_bwd]):
    print('Analyzing for',uma_trainers[i])
    fall_ser = pd.Series()
    no_fall_ser = pd.Series()
    fall_ser['TrainingSet'] = uma_trainers[i]
    fall_ser['Class'] = 'Caida'
    no_fall_ser['TrainingSet'] = uma_trainers[i]
    no_fall_ser['Class'] = 'NoCaida'
    for class_name, class_base, class_opts in classifiers_uma:
        print(class_name, class_base, class_opts)
        cls = Classifier(classname=class_base, options=class_opts)
        cls.build_classifier(df)
        evaluation = Evaluation(sim)
        evl = evaluation.test_model(cls, sim)   
        fall_ser[class_name] = round(evaluation.f_measure(0),3)
        no_fall_ser[class_name] = round(evaluation.f_measure(1),3)
        if (REDUCED_PRINT):
            print("Caida fMeasure: " + str(round(evaluation.f_measure(0),3)))
            print("NoCaida fMeasure: " + str(round(evaluation.f_measure(1),3)))
        else: 
            print(evaluation.summary())
            print(evaluation.class_details())

    resumen_df = resumen_df.append([fall_ser,no_fall_ser])


for i, df in enumerate([up_sdwd, up_fwd_knees, up_fwd_hands, up_bwd]):
    print('Analyzing for',up_trainers[i])
    fall_ser = pd.Series()
    no_fall_ser = pd.Series()
    fall_ser['TrainingSet'] = up_trainers[i]
    fall_ser['Class'] = 'Caida'
    no_fall_ser['TrainingSet'] = up_trainers[i]
    no_fall_ser['Class'] = 'NoCaida'
    for class_name, class_base, class_opts in classifiers_up:
        print(class_name, class_base, class_opts)
        cls = Classifier(classname=class_base, options=class_opts)
        cls.build_classifier(df)
        evaluation = Evaluation(sim)
        evl = evaluation.test_model(cls, sim)  
        fall_ser[class_name] = round(evaluation.f_measure(0),3)
        no_fall_ser[class_name] = round(evaluation.f_measure(1),3)  
        if (REDUCED_PRINT):
            print("Caida fMeasure: " + str(round(evaluation.f_measure(0),3)))
            print("NoCaida fMeasure: " + str(round(evaluation.f_measure(1),3)))
        else: 
            print(evaluation.summary())
            print(evaluation.class_details())
            
    resumen_df = resumen_df.append([fall_ser,no_fall_ser])
    
    
display(resumen_df)
```
